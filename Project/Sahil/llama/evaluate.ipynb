# English to Hindi

```python
import evaluate
from datasets import load_dataset
from tqdm import tqdm
```

```python
metric = evaluate.load("sacrebleu")
valid_dataset = load_dataset("cfilt/iitb-english-hindi")
```

```python
original_dataset = [valid_dataset['validation'][i]['translation']['hi'] for i in range(len(valid_dataset['validation']))]
```

```python
len(original_dataset)
```

```python
with open('/data5/home/sahilm/NLP_Project/llama/output_eng_to_hin_13b.txt', 'r') as file:
    output = file.readlines()
```

```python
len(output)
```

```python
predicted = []

for i in range(len(output)):
    # print(output[i])
    if output[i].startswith("Translated Hindi:"):
        predicted.append(output[i].split("Translated Hindi:")[1].strip())
   
```

```python
len(predicted)
```

```python
for i in range(len(predicted)):
    print(predicted[i])
    print(original_dataset[i])
    print('-'*35)
```

```python
total_bleu = 0.0
for i in range(len(predicted)):
    total_bleu += metric.compute(predictions=[predicted[i]], references=[[original_dataset[i]]])['score']

print("Final BLEU Score: ", total_bleu/len(predicted))
```

# Hindi to English

```python
original_dataset = [valid_dataset['validation'][i]['translation']['en'] for i in range(len(valid_dataset['validation']))]
```

```python
len(original_dataset)
```

```python
with open('/data5/home/sahilm/NLP_Project/open_hathi/open_hathi_hin_to_eng.txt', 'r') as file:
    output = file.readlines()
```

```python
len(output)
```

```python
predicted = []
actual = []

flag = True
for i in range(len(output)):
    # print(output[i])
    if flag and output[i].startswith("[English]:"):
        predicted.append(output[i].split("[English]:")[1].strip())
        flag = False
    if output[i].startswith("[Actual English]:"):
        actual.append(output[i].split("[Actual English]:")[1].strip())
        flag = True
```

```python
print(len(predicted))
print(len(actual))
```

```python
actual==original_dataset
```

```python
for i in range(len(original_dataset)):
    if original_dataset[i]!=actual[i]:
        print(original_dataset[i])
        print(actual[i])
```

```python
for i in range(len(predicted)):
    print(predicted[i])
    print(actual[i])
    print('-'*35)
```

```python
total_bleu = 0.0
for i in range(len(predicted)):
    total_bleu += metric.compute(predictions=[predicted[i]], references=[[actual[i]]])['score']

print("Final BLEU Score: ", total_bleu/len(predicted))
```

```python
total_bleu = 0.0
for i in range(len(predicted)):
    total_bleu += metric.compute(predictions=[predicted[i]], references=[[original_dataset[i]]])['score']

print("Final BLEU Score: ", total_bleu/len(predicted))
```
