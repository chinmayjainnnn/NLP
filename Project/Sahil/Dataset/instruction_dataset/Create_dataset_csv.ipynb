```python
from datasets import load_dataset

with open("/data5/home/sahilm/NLP_Project/Dataset/instruction_dataset/English.txt",'r') as file1:
    english_sentences = file1.readlines()


with open("/data5/home/sahilm/NLP_Project/Dataset/instruction_dataset/Hindi.txt",'r') as file2:
    hindi_sentences = file2.readlines()
```

```python
DEFAULT_SYSTEM_PROMPT = """
Translate from English to Hindi
""".strip()
```

```python
def generate_training_prompt(
    input: str, response: str, system_prompt: str = DEFAULT_SYSTEM_PROMPT
) -> str:
    return f"""Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request. 
    

### Instruction:
{system_prompt}


### Input:
{input.strip()}


### Response:
{response.strip()}
""".strip()
```

```python
import json

def generate_text(input, response):
 
    return {
        "input": input,
        "response": response,
        "text": generate_training_prompt(input, response),
    }
```

```python
import csv
```

```python
def read_data_from_txt(i):
        # Assuming data is separated by tabs (\t)
        english, hindi, text = english_sentences[i], hindi_sentences[i], generate_text(english_sentences[i],hindi_sentences[i])['text']
        return [english, hindi, text]
```

```python
def write_to_csv(csv_file):
    with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        data = []
        for i in range(len(english_sentences)):
            # print(read_data_from_txt(i))
            data.append(read_data_from_txt(i))
        print(data)
        writer.writerow(["English", "Hindi", "Text"])
        writer.writerows(data)
```

```python
csv_file = "data.csv"
write_to_csv(csv_file)
```
