```python
import re    

# @sahil lathiya set file path before running
```

```python
def is_hindi_char(char):
    # Range of Unicode characters for Hindi script
    return 0x0900 <= ord(char) <= 0x097F
```

```python
pattern = r'Translated Hindi: (.*)'

with open('/raid/home/chinmayjain/NLP_project/Output/output_eng_to_hin_7b.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

extracted_lines = []
```

```python
indx_of_lines = []
cnt = 0
for line in lines:
    match = re.match(pattern, line)
    if match:
        hindi_text = match.group(1).strip()
        hindi_text_check = hindi_text[:10]  # Take only the first 5 characters
        # Check if the extracted initial 5 characters contain only Hindi characters
        temp=0
        for char in hindi_text_check:
            if not (is_hindi_char(char) or char.isdigit()  or char.isspace() or char in ['।', ',', '(', ')'] ):
                temp=1
        if temp==0:
            extracted_lines.append(hindi_text)
            indx_of_lines.append(cnt)
        cnt+=1
        
```

```python
# Print the extracted lines
for extracted_line in extracted_lines:
    print(extracted_line)
```
